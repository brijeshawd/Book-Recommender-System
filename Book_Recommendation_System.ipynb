## **Popularity based Recommendation System**

Steps:

1. Using avg rating concept on most rated books:
2. Most rated books: where number of rating is more
3. Most rated books: group by book title and count no of ratings on that book
4. avg rating: group by book title and find mean of ratings on that book

```python
ratings_with_name=ratings.merge(books, on="ISBN")
```

```python
ratings_with_name
```

```python
# finding number of rating and avg rating per book
rating_info=ratings_with_name.groupby(["Book-Title"]).agg({"Book-Rating":["count","mean"]}).reset_index()
rating_info.columns=["Book-Title","num_ratings","mean_ratings"]
rating_info
```

```python
# Fetching rows where no. of rating is greater than 250 to get popular books
popular_df=rating_info[rating_info["num_ratings"]>=250].sort_values(by="mean_ratings", ascending=False)
# Fetching Top 50 Books info where max people rated books with greatest avg rating
popular_df=popular_df.head(50)
popular_df
```

```python
# need author name, publisher name and image for poster so we will merge this to books dataframe

popular_df.merge(books, on="Book-Title")
```

```python
# there are multiple duplicates present and hence total rows becomes 196, so we have to add drop duplicates
popular_df.merge(books, on="Book-Title").drop_duplicates("Book-Title").shape
```

```python
popular_df=popular_df.merge(books, on="Book-Title").drop_duplicates("Book-Title")[["Book-Title","Book-Author","Image-URL-M","num_ratings","mean_ratings"]]
```

## **Collaborative Filtering Based Recommender System**

```python
ratings_with_name.head()
```

```python
ratings_with_name.groupby(["User-ID"]).count()["Book-Rating"]
```

```python
# fetching user id of those users who have given rating on atleast 200 Books
temp=ratings_with_name.groupby(["User-ID"]).count()["Book-Rating"].reset_index()
imp_users=temp[temp["Book-Rating"]>200]["User-ID"]
imp_users
```

```python
# these 811 users given rating leading to these 4Lakh rows
# filtered based on users
filtered_ratings=ratings_with_name[ratings_with_name["User-ID"].isin(imp_users)]
```

```python
# now we have to apply filter based on books
# finding books which were given more than 50 ratings and get the book title
temp=filtered_ratings.groupby(["Book-Title"]).count()["Book-Rating"].reset_index()
imp_books=temp[temp["Book-Rating"]>50]["Book-Title"]
```

```python
final_ratings=filtered_ratings=filtered_ratings[filtered_ratings["Book-Title"].isin(imp_books)]
```

```python
# item user similarity matrix 
book_user_matrix=final_ratings.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
book_user_matrix.fillna(0,inplace=True)
book_user_matrix
```

```python
# finding cosine similarity
from sklearn.metrics.pairwise import cosine_similarity

similarity_score=cosine_similarity(book_user_matrix)
```

```python
similarity_score.shape
```

**Steps:**
1. fetch book index
2. find the value(row) of that index in similarity score
3. use enumerate and sort in descending order to find the index of largest value
4. loop top 5 index and find book title using these index

```python
def recommend(book_name):
  #1. fetch book index
  book_index=np.where(book_user_matrix.index==book_name)[0][0]
  #2. find the value(row) of that index in similarity score
  distance=similarity_score[book_index]
  #3. use enumerate and sort in descending order to find the index of largest value
  similar_items=sorted(list(enumerate(similarity_score[0])), reverse=True, key=lambda x: x[1])[1:6]
  #4. loop top 5 index and find book title using these index
  for i in similar_items:
    print(book_user_matrix.index[i[0]])
```

```python
recommend("1984")
```
